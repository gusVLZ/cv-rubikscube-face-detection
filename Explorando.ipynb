LIBRARIES

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

LOAD VIDEO AND CAPTURE A FRAME PER SEC

```python
cap = cv.VideoCapture("video/VOD_VALORANT_SAMPLE.mp4")
fps = cap.get(cv.CAP_PROP_FPS)
sec = 0

while(cap.isOpened()):
    cap.set(cv.CAP_PROP_POS_FRAMES,sec*fps+1)
    ret, frame = cap.read()

    if ret == False:
        break

    cv.imwrite('outputs/vod1/Frame'+str(sec)+'.jpg', frame)
    sec+=1

cap.release()

```

LOAD IMAGE CUT AND PLAYS WITH ITS COLOR
