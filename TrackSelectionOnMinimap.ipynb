```python
def roundColors(arr):
    for i, x in enumerate(arr):
        for k, y in enumerate(x):
            rounded = math.floor(y/10)*10
            arr[i][k] = rounded
    return arr
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

cap = cv2.VideoCapture("video/ROUND_MINIMAP.mp4")
cap.set(cv2.CAP_PROP_POS_FRAMES,1)
(grabbed, frame) = cap.read()
frame = cv2.resize(frame, None, None, 3, 3)
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
roi = cv2.selectROI("Select region", frame, False, False)
template = gray[roi[1]:roi[1]+roi[3],roi[0]:roi[0]+roi[2]]
h, w = template.shape

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.resize(frame, None, None, 3, 3)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    res = cv2.matchTemplate(gray, template, cv2.TM_CCOEFF_NORMED)
    threshold = 0.8

    loc = np.where( res >= threshold)  

    for pt in zip(*loc[::-1]): 
        cv2.rectangle(frame, pt, (pt[0] + w, pt[1] + h), (0, 0, 255), 2)

    cv2.imshow("Matched image", frame)
    cv2.imshow("Matched gray", gray)
    key = cv2.waitKey(0)
    if key == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
